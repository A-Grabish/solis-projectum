```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def clean_trim_year(value):
    value = str(value)
    result = value[6:]
    return result
```

```python
# max temp

# Raw data file(s)
csv_file = "max_raw.csv"
max_raw = pd.read_csv(csv_file, 
                      sep    = "  ", 
                      engine = 'python', 
                      names  = ['Year', 'Jan', 'Feb', 'March', 'April', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec'])
```

```python
max_nation = max_raw[(max_raw['Year'] < 1110000000) & (max_raw['Year'] > 1100000000)]
max_nation_df = pd.DataFrame(max_raw[(max_raw['Year'] < 1110000000) & (max_raw['Year'] > 1100000000)])
```

```python
# clean year col and reset index
#max_nation['Year'] = max_nation['Year'] - 1100270000

max_nation_df['Year'] = max_nation_df['Year'].apply(clean_trim_year)
max_nation_df.head()
```
